# Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
df = pd.read_csv("istherecorrelation.csv", delimiter=";")

# Replace commas with dots for WO [x1000] column and convert to float
df["WO [x1000]"] = df["WO [x1000]"].str.replace(",", ".").astype(float)


corr = df[["NL Beer consumption [x1000 hectoliter]", "WO [x1000]"]].corr().iloc[0,1]

# Plot the data in same graph with seperate y-axes
fig, ax1 = plt.subplots()
sns.lineplot(x="Year", y="NL Beer consumption [x1000 hectoliter]", label="Beer Consumption",data=df, ax=ax1, color="orange")
ax1.set_ylabel("Beer Consumption [x1000 hectoliter]")
ax1.legend(loc="upper left")

ax2 = ax1.twinx()
sns.lineplot(x="Year", y="WO [x1000]", label="Number of Students", data=df, ax=ax2)
ax2.set_ylabel("Number of Students [x1000]")
ax2.legend(loc="upper right")

plt.text(0.8, 0.9, f"Correlation WOxBeer: {corr:.2f}", transform=plt.gcf().transFigure, fontsize=10, ha='center')

plt.show()
plt.savefig("beer_vs_students.png")
```
